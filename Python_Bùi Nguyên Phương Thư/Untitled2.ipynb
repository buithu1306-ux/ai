```python
#câu 14
#a
import random
n = int (input("nhap số cột: "))
m = int (input("nhap số dòng: "))
a=[]
for i in range(m):
  dong =[]
  for j in range(n):
    snn = random.randint(1,100)
    dong.append(snn)
  a.append(dong)
print("ma trận vừa tạo là: ")
for dong in a:
  print(dong)

#b
k= int(input("nhập dòng muốn xuất: "))
if 0<= k<=m:
  print(f"các phần tử thuộc dòng {k}: ",a[k-1])

#c
k= int(input("nhập cột muốn xuất: "))
print(f"các phần tử thuộc cột {k}: ", [row[k-1] for row in a])

#d
max_sum=-1
max_row= 0
for i in range(m):
  tong = sum(a[i])
if tong > max_sum:
  max_sum= tong
  max_row= i
print(f"dòng có tổng lớn nhất là {max_row} với tổng là: ", max_sum)

#e
min_tich = 1
for i in range(m):
    min_tich *= a[i][0]
cot_min = 1
for j in range(1, n):
    tich_hien_tai = 1
    for i in range(m):
        tich_hien_tai *= a[i][j]
    if tich_hien_tai < min_tich:
        min_tich = tich_hien_tai
        cot_min = j
print(f"Cột có tích nhỏ nhất là {cot_min} với tích = {min_tich}")

#f
print("Các phần tử thuộc dong chẳn: ")
for i in range(m):
    if  i%2 ==0:
       print(f"các phần tử thuộc dòng {i}: ",a[i-1])
print("Các phần tử thuộc cột lẻ:")
for j in range(n):
    if j%2 !=0:
       print(f"các phần tử thuộc dòng {j}: ",a[j-1])

#g
tong, dem = 0, 0
for i in range(1, m, 2):
    for x in a[i]:
        if x % 2 == 0:
            tong += x
            dem += 1
    print(f"TBC chẵn dòng lẻ {i}:", tong/dem if dem > 0 else 0)

#h
tong, dem = 0, 0
for i in range(m):
  for j in range(n):
    if i==0 or i==m-1 or j == 0 or j == n - 1:
      tong += a[i][j]
      dem += 1
print("TBC biên: ", tong/dem if dem > 0 else 0)

#i
tong, dem = 0, 0
for i in range(m):
  for j in range(n):
    if i!=0 or i!=m-1 or j != 0 or j != n - 1:
      tong += a[i][j]
      dem += 1
print("TBC các phần tử không thuộc biên: ", tong/dem if dem > 0 else 0)
```

```python
#câu 15
from datetime import date
# a
class SINHVIEN:
    def __init__(self, ma, ten, nam_sinh, dtb):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.dtb = dtb

# b.
n = int(input("Nhập số lượng sinh viên: "))
ds_sv = []
for i in range(n):
    print(f"--- Nhập thông tin sinh viên thứ {i+1} ---")
    ma = input("Mã sinh viên: ")
    ten = input("Họ tên: ")
    nam_sinh = int(input("Năm sinh: "))
    dtb = float(input("Điểm trung bình: "))
    sv_moi = SINHVIEN(ma, ten, nam_sinh, dtb)
    ds_sv.append(sv_moi)

# c
len_lop = sum(1 for sv in ds_sv if sv.dtb >= 5)
print("Số sinh viên đủ điều kiện lên lớp:",len_lop)

# d
nam_nay = date.today().year
sv_20_tuoi = [sv.ten for sv in ds_sv if nam_nay - sv.nam_sinh >= 20]
print("SV đủ 20 tuổi:", sv_20_tuoi)

# e
sv_dh = [sv for sv in ds_sv if sv.ma[2:4] == "DH"]
print(f"Số SV hệ Đại học: {len(sv_dh)}")

# f
count_lan = sum(1 for sv in ds_sv if sv.ten.split()[-1] == "Lan")
print(f"Số người tên Lan: {count_lan}")

# g
count_phan = sum(1 for sv in ds_sv if sv.ten.startswith("Phan"))
print(f"Số người họ Phan: {count_phan}")
```

```python
#Câu 1 - BTVN
def TronMang(a, b):
    kq = []
    i = 0
    j = 0
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            kq.append(a[i])
        else:
            kq.append(b[j])
        i += 1
        j += 1
    while i < len(a):
        kq.append(a[i])
        i += 1
    while j < len(b):
        kq.append(b[j])
        j += 1
    return kq
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print("Mảng kết quả:", TronMang(a, b))
```

```python
#bài 2 - BTVN
import numpy as np

n = 4
a = np.random.randint(-10, 10, (n, n))
print("Ma trận gốc:\n", a)

# a
tong = 0
for i in range(n):
      for j in range(n):
         if i + j <= n - 1:
            tong += a[i][j]
print("Tổng các phần tử nằm trên tam giác dưới:", tong)

# b
print("Ma trận sau khi chuyển đổi:\n", np.abs(a))

# c
x= int(input("nhập sô nguyên: "))
a[a % 2 == 0] = x
print(f"Ma trận sau khi thay thế số chẳn băng {x}:\n", a)

# d
print("Ma trận có phải toàn chẳn không? ",np.all(a % 2 == 0))

# e
print("Ma trận có đối xứng không? ",np.allclose(a, a.T))

# f
diag_main = np.diag(a)
is_main_inc = np.all(np.diff(diag_main) > 0)
print("Đường chéo chính có tăng dần không? ",is_main_inc)

# g
for i in range(n):
    for j in range(n - 1 - i, n):
        print(a[i][j], end=" ")
    print()

# h
diag_sub = np.diag(np.fliplr(a))
is_sub_dec = np.all(np.diff(diag_sub) < 0)
print("đường chéo phụ có giảm dần không? ",is_sub_dec)
```
