```python
# BTVN 2
from collections import deque

class GraphTSP:
    def __init__(self, adjac_lis, all_nodes):
        self.adjac_lis = adjac_lis
        self.all_nodes = sorted(all_nodes)

    def get_neighbors(self, v_tuple):
        curr_node, visited = v_tuple
        neighbors = []

        if curr_node in self.adjac_lis:
            for (m, weight) in self.adjac_lis[curr_node]:
                if m not in visited:
                    new_visited = tuple(sorted(list(visited) + [m]))
                    neighbors.append(((m, new_visited), weight))
                elif len(visited) == len(self.all_nodes) and m == self.all_nodes[0]:
                    new_visited = tuple(sorted(list(visited) + ["FINISH"]))
                    neighbors.append(((m, new_visited), weight))
        return neighbors

    def h(self, v_tuple):
        curr_node, visited = v_tuple
        if "FINISH" in visited:
            return 0

        remaining_count = len(self.all_nodes) - len(visited)
        return remaining_count * 1

    def a_star(self, start_node):
        start_state = (start_node, (start_node,))

        open_lst = set([start_state])
        closed_lst = set([])

        poo = {}
        poo[start_state] = 0

        par = {}
        par[start_state] = start_state

        while len(open_lst) > 0:
            n = None
            for v in open_lst:
                if n == None or poo[v] + self.h(v) < poo[n] + self.h(n):
                    n = v

            if n == None:
                print('Path does not exist')
                return None

            curr_city, visited = n
            if len(visited) == len(self.all_nodes) + 1 and curr_city == start_node:
                reconst_path = []
                temp_n = n
                while par[temp_n] != temp_n:
                    reconst_path.append(temp_n[0])
                    temp_n = par[temp_n]
                reconst_path.append(start_node)
                reconst_path.reverse()

                print('TSP Path found: {}'.format(reconst_path))
                print('Total Cost: {}'.format(poo[n]))
                return reconst_path

            for (m_state, weight) in self.get_neighbors(n):
                if m_state not in open_lst and m_state not in closed_lst:
                    open_lst.add(m_state)
                    par[m_state] = n
                    poo[m_state] = poo[n] + weight
                else:
                    if poo[m_state] > poo[n] + weight:
                        poo[m_state] = poo[n] + weight
                        par[m_state] = n
                        if m_state in closed_lst:
                            closed_lst.remove(m_state)
                            open_lst.add(m_state)

            open_lst.remove(n)
            closed_lst.add(n)

        print('Path does not exist')
        return None

adjac_lis = {
    'A': [('B', 10), ('C', 15), ('D', 20)],
    'B': [('A', 10), ('C', 35), ('D', 25)],
    'C': [('A', 15), ('B', 35), ('D', 30)],
    'D': [('A', 20), ('B', 25), ('C', 30)]
}
all_cities = ['A', 'B', 'C', 'D']

graph_tsp = GraphTSP(adjac_lis, all_cities)
graph_tsp.a_star('A')
```
