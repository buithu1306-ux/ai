```python
#Bài 3. Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.
import copy
import math
import numpy as np

X = "X"
O = "O"
EMPTY = None
WIN_CONDITION = 5 # Mặc định 5 quân liên tiếp là thắng cho N >= 5

def initial_state(n):
    return [[EMPTY for _ in range(n)] for _ in range(n)]

def player(board):
    count = 0
    for i in board:
        for j in i:
            if j: count += 1
    return O if count % 2 != 0 else X

def actions(board):
    res = set()
    n = len(board)
    for i in range(n):
        for j in range(n):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    curr_player = player(board)
    result_board = [list(row) for row in board]
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def check_win(board, p):
    n = len(board)
    # Check ngang, dọc, chéo
    for i in range(n):
        for j in range(n):
            if board[i][j] == p:
                # Ngang
                if j + WIN_CONDITION <= n and all(board[i][j+k] == p for k in range(WIN_CONDITION)): return True
                # Dọc
                if i + WIN_CONDITION <= n and all(board[i+k][j] == p for k in range(WIN_CONDITION)): return True
                # Chéo xuôi
                if i + WIN_CONDITION <= n and j + WIN_CONDITION <= n and all(board[i+k][j+k] == p for k in range(WIN_CONDITION)): return True
                # Chéo ngược
                if i + WIN_CONDITION <= n and j - WIN_CONDITION >= -1 and all(board[i+k][j-k] == p for k in range(WIN_CONDITION)): return True
    return False

def winner(board):
    if check_win(board, X): return X
    if check_win(board, O): return O
    return None

def terminal(board):
    if winner(board) is not None: return True
    return all(all(cell != EMPTY for cell in row) for row in board)

def heuristic_utility(board):
    """Đánh giá trạng thái khi chưa kết thúc game (Heuristic)"""
    # Hàm này ước lượng điểm dựa trên ưu thế vị trí, đơn giản nhất là 0 nếu chưa thắng
    # Trong các bài toán NxN, đây là nơi cài đặt logic thông minh cho AI
    win_val = winner(board)
    if win_val == X: return 1000
    if win_val == O: return -1000
    return 0

def maxValue(state, depth, alpha, beta):
    if terminal(state) or depth == 0:
        return heuristic_utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action), depth - 1, alpha, beta))
        alpha = max(alpha, v)
        if beta <= alpha: break # Cắt tỉa để chạy được NxN
    return v

def minValue(state, depth, alpha, beta):
    if terminal(state) or depth == 0:
        return heuristic_utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action), depth - 1, alpha, beta))
        beta = min(beta, v)
        if beta <= alpha: break
    return v

def minimax(board):
    curr_player = player(board)
    n = len(board)
    # Giới hạn độ sâu tùy theo N để tránh đứng máy
    depth_limit = 3 if n <= 5 else 2

    best_move = None
    if curr_player == X:
        v = -math.inf
        for action in actions(board):
            val = minValue(result(board, action), depth_limit, -math.inf, math.inf)
            if val > v:
                v = val
                best_move = action
    else:
        v = math.inf
        for action in actions(board):
            val = maxValue(result(board, action), depth_limit, -math.inf, math.inf)
            if val < v:
                v = val
                best_move = action
    return best_move

if __name__ == "__main__":
    n = int(input("Nhập kích thước NxN: "))
    WIN_CONDITION = 3 if n == 3 else (4 if n == 4 else 5)
    board = initial_state(n)

    user_choice = input("Chọn X hoặc O: ").upper()
    user = user_choice
    ai = O if user == X else X

    while not terminal(board):
        curr_p = player(board)
        print(f"\nLượt của {curr_p}:")

        if curr_p == user:
            r = int(input(f"Nhập hàng (0-{n-1}): "))
            c = int(input(f"Nhập cột (0-{n-1}): "))
            if board[r][c] == EMPTY:
                board = result(board, (r, c))
            else:
                print("Ô đã có người đi!")
                continue
        else:
            print("AI đang tính toán...")
            move = minimax(board)
            board = result(board, move)

        print(np.array(board))

    win = winner(board)
    if win: print(f"Kết thúc! {win} thắng.")
    else: print("Kết thúc! Hòa.")
```

```python
#Bài 6. Xây dựng giao diện đồ họa cho trò chơi tictactoe.
import math
from IPython.display import display, clear_output
import ipywidgets as widgets

X = "X"
O = "O"
EMPTY = None

class TicTacToeColab:
    def __init__(self):
        self.board = [[EMPTY for _ in range(3)] for _ in range(3)]
        self.user = X
        self.ai = O
        self.buttons = []
        self.output = widgets.Output()
        self.grid = self.create_grid()

    def create_grid(self):
        items = []
        for i in range(3):
            row_buttons = []
            for j in range(3):
                btn = widgets.Button(description="", layout=widgets.Layout(width='60px', height='60px'))
                btn.on_click(self.make_on_click(i, j))
                row_buttons.append(btn)
            items.append(widgets.HBox(row_buttons))
        return widgets.VBox(items)

    def make_on_click(self, r, c):
        def on_click(btn):
            if self.board[r][c] is EMPTY and not self.terminal():
                self.update_cell(r, c, self.user)
                if not self.terminal():
                    move = self.minimax_decision()
                    if move:
                        self.update_cell(move[0], move[1], self.ai)

                self.check_game_over()
        return on_click

    def update_cell(self, r, c, player):
        self.board[r][c] = player
        btn = self.grid.children[r].children[c]
        btn.description = player
        btn.disabled = True
        btn.style.button_color = '#d1e7ff' if player == X else '#f8d7da'

    def check_game_over(self):
        if self.terminal():
            res = self.winner(self.board)
            with self.output:
                clear_output()
                if res:
                    print(f"🎉 KẾT THÚC: {res} THẮNG!")
                else:
                    print("🤝 KẾT THÚC: HÒA!")

    def winner(self, board):
        for i in range(3):
            if board[i][0] == board[i][1] == board[i][2] != EMPTY: return board[i][0]
            if board[0][i] == board[1][i] == board[2][i] != EMPTY: return board[0][i]
        if board[0][0] == board[1][1] == board[2][2] != EMPTY: return board[0][0]
        if board[0][2] == board[1][1] == board[2][0] != EMPTY: return board[0][2]
        return None

    def terminal(self):
        return self.winner(self.board) is not None or all(all(c != EMPTY for c in r) for r in self.board)

    def utility(self, board):
        res = self.winner(board)
        return 1 if res == X else (-1 if res == O else 0)

    def minimax_decision(self):
        best_val = math.inf
        best_move = None
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == EMPTY:
                    self.board[i][j] = self.ai
                    val = self.max_value(self.board)
                    self.board[i][j] = EMPTY
                    if val < best_val:
                        best_val, best_move = val, (i, j)
        return best_move

    def max_value(self, board):
        if self.terminal(): return self.utility(board)
        v = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = X
                    v = max(v, self.min_value(board))
                    board[i][j] = EMPTY
        return v

    def min_value(self, board):
        if self.terminal(): return self.utility(board)
        v = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = O
                    v = min(v, self.max_value(board))
                    board[i][j] = EMPTY
        return v

# Chạy trò chơi
game = TicTacToeColab()
display(game.grid, game.output)
```
