```python
#Bài 3. Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.
import copy
import math
import numpy as np

X = "X"
O = "O"
EMPTY = None
WIN_CONDITION = 5 # Mặc định 5 quân liên tiếp là thắng cho N >= 5

def initial_state(n):
    return [[EMPTY for _ in range(n)] for _ in range(n)]

def player(board):
    count = 0
    for i in board:
        for j in i:
            if j: count += 1
    return O if count % 2 != 0 else X

def actions(board):
    res = set()
    n = len(board)
    for i in range(n):
        for j in range(n):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    curr_player = player(board)
    result_board = [list(row) for row in board]
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def check_win(board, p):
    n = len(board)
    # Check ngang, dọc, chéo
    for i in range(n):
        for j in range(n):
            if board[i][j] == p:
                # Ngang
                if j + WIN_CONDITION <= n and all(board[i][j+k] == p for k in range(WIN_CONDITION)): return True
                # Dọc
                if i + WIN_CONDITION <= n and all(board[i+k][j] == p for k in range(WIN_CONDITION)): return True
                # Chéo xuôi
                if i + WIN_CONDITION <= n and j + WIN_CONDITION <= n and all(board[i+k][j+k] == p for k in range(WIN_CONDITION)): return True
                # Chéo ngược
                if i + WIN_CONDITION <= n and j - WIN_CONDITION >= -1 and all(board[i+k][j-k] == p for k in range(WIN_CONDITION)): return True
    return False

def winner(board):
    if check_win(board, X): return X
    if check_win(board, O): return O
    return None

def terminal(board):
    if winner(board) is not None: return True
    return all(all(cell != EMPTY for cell in row) for row in board)

def heuristic_utility(board):
    """Đánh giá trạng thái khi chưa kết thúc game (Heuristic)"""
    # Hàm này ước lượng điểm dựa trên ưu thế vị trí, đơn giản nhất là 0 nếu chưa thắng
    # Trong các bài toán NxN, đây là nơi cài đặt logic thông minh cho AI
    win_val = winner(board)
    if win_val == X: return 1000
    if win_val == O: return -1000
    return 0

def maxValue(state, depth, alpha, beta):
    if terminal(state) or depth == 0:
        return heuristic_utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action), depth - 1, alpha, beta))
        alpha = max(alpha, v)
        if beta <= alpha: break # Cắt tỉa để chạy được NxN
    return v

def minValue(state, depth, alpha, beta):
    if terminal(state) or depth == 0:
        return heuristic_utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action), depth - 1, alpha, beta))
        beta = min(beta, v)
        if beta <= alpha: break
    return v

def minimax(board):
    curr_player = player(board)
    n = len(board)
    # Giới hạn độ sâu tùy theo N để tránh đứng máy
    depth_limit = 3 if n <= 5 else 2

    best_move = None
    if curr_player == X:
        v = -math.inf
        for action in actions(board):
            val = minValue(result(board, action), depth_limit, -math.inf, math.inf)
            if val > v:
                v = val
                best_move = action
    else:
        v = math.inf
        for action in actions(board):
            val = maxValue(result(board, action), depth_limit, -math.inf, math.inf)
            if val < v:
                v = val
                best_move = action
    return best_move

if __name__ == "__main__":
    n = int(input("Nhập kích thước NxN: "))
    WIN_CONDITION = 3 if n == 3 else (4 if n == 4 else 5)
    board = initial_state(n)

    user_choice = input("Chọn X hoặc O: ").upper()
    user = user_choice
    ai = O if user == X else X

    while not terminal(board):
        curr_p = player(board)
        print(f"\nLượt của {curr_p}:")

        if curr_p == user:
            r = int(input(f"Nhập hàng (0-{n-1}): "))
            c = int(input(f"Nhập cột (0-{n-1}): "))
            if board[r][c] == EMPTY:
                board = result(board, (r, c))
            else:
                print("Ô đã có người đi!")
                continue
        else:
            print("AI đang tính toán...")
            move = minimax(board)
            board = result(board, move)

        print(np.array(board))

    win = winner(board)
    if win: print(f"Kết thúc! {win} thắng.")
    else: print("Kết thúc! Hòa.")
```
